##### Biosynthetic pathway/genes
In the obligate fermentative bacterium Zymomonas mobilis which is one of few prokaryotes can generate ethanol as the predominant fermentative product. Yeasts such as Saccharomyces cerevisiae are also widely used for the production of ethanol. Two enzymes, pyruvate decarboxylase (PDC) and alcohol dehydrogenase (ADHE) catalyze the major pathway for ethanol synthesis in Z. mobilis and S. cerevisiae (Fig.1.1)(Danner and Braun, 1999). PDC catalyzes the no-oxidative decarboxylation of pyruvate, which produces acetaldehyde and CO2. ADH then converts acetaldehyde to ethanol(Deng and Coleman, 1999). 


The lactic acid bacteria used in this report has an innate high tolerance ethanol and are frequently found in bio-ethanol plants as contaminants. In Lactococcus lactis, pyruvate formed in glycolysis is converted to acetyl-CoA by pyruvate dehydrogenase complex (PDHC), and then bifunctional native alcohol dehydrogenase (ADHE) which can use either acetyl-CoA or acetaldehyde as substrate (Fig.1.2). This is different from ADHE in Zymomonas mobilis and Saccharomyces cerevisiae which only uses acetaldehyde as the substrate(Solem et al., 2013).

##### General advantages
Lactic acid bacteria (LAB) are a heterogeneous group of microorganisms able to convert carbohydrates into lactic acid. As the long tradition of safe use (GRAS microorganism), they are applied worldwide in the industrial manufacture of fermented food products. Among LAB, Lactococcus lactis is, by far, the most extensively studied with respect to its physiology, metabolic pathways and regulatory mechanisms, and its genome was the first LAB genome to be completely sequenced and annotated, which laid a foundation for us to execute genetic manipulations and analysis. Meanwhile, because of the properties of its high glycolytic flux, ability to metabolize a broad range of carbohydrates, well-characterized metabolic network, high ethanol and acid tolerance and ease of genetic manipulation, Lactococcus lactis has its own potential to be used as cell-factories in bioreactors for the in situ production of compounds that contribute to the flavor, texture or health benefits of foods(Liu et al., 2016)(Solem et al., 2013).


##### General disadvantages
<i> L. lactis </i> is a facultative anaerobic bacterium. It lacks a functional respiratory chain. ATP is only formed through substrate level phosphorylation. That leads to specific nutritional growth requirements is called for, which could perhaps make it less attractive for some industrial applications. 

Another potential drawback of using L. lactis as a cell factory is most of the carbon flux is directed to lactate (homo-lactic fermentation). For example, during growth on glucose, only about 5% of the metabolized sugar is converted into biomass.  That implies that L. lactis have a very limited biosynthetic capacity(Oliveira et al., 2005).


##### Suitability of the cell factory for the product
Currently, there is a growing demand for liquid fuels that can be produced in a sustainable manner from renewable raw materials. Microbial production of fuel compounds has gained a lot of attention and focus has mostly been on developing bio-processes involving non-food plant biomass feedstocks.  

However, The high cost of the enzymes needed to degrade such feedstocks into its constituent sugars as well as problems due to various inhibitors generated in pretreatment are two challenges that have to be addressed if cost-effective processes are to be established. Cheese whey, which referred as a cheap abundant feedstock, can be used to produce ethanol alternatively. And in this acid and lactose-rich substrate,<i> L. lactis </i> will be a good potential ethanol production host organism, which will not only alleviate an existing environmental problem but also produce useful value-added chemical in a large industrail waste stream(Liu et al., 2016). 

## 2. Problem definition <a class="anchor" id="Problem definition"></a>

To forsake modern society’s dependency on limited, non-renewable oil-reserves, recent approaches have aimed at constructing microbial hosts for the production of bio- alcohols as alternative fuels. 

Lactic acid bacteria currently used extensively by the dairy industry have a superior tolerance towards short-chain alcohols, which makes them interesting targets for use in future bio-refineries. But a major hurdle limiting the yield and titers of microbial bio-alcohol production is the toxicity of these compounds(Vinuselvi et al., 2011). 

As ethanol and butanol are both short-chain alcohols there are similarities in how microbes respond to them. Through their hydrophobic nature, short-chain alcohols disrupt the phospholipid components of the cell membrane thereby increasing membrane fluidity and destroying the integrity of the cell. Besides, ethanol stress, the most studied alcohol stress, has been shown to increase membrane permeability, allowing for the free passage of ions and charged molecules. And high concentrations of ethanol and butanol have been shown to interfere with the pH gradient of cell membranes, thereby increasing acid sensitivity in different strains. 

Specifically, research revealed that while high concentrations of ethanol and hexanol were cytostatic to <i>L. lactis</i>, high concentrations of butanol were cytotoxic through disruption of the cell membrane. Therefore, adaptive laboratory evolution and efficiently engineering metabolism has been initiated to provide ethanol producing L. lactis strain(Hviid et al., 2017).

## 3. Selection and assessment of existing GSM <a class="anchor" id="Selection and assessment of existing GSM"></a>

Several LAB-model organisms such as Lactococcus lactis IL1403, Lactobacillus plantarum WCFS1 and Streptococcus thermophilus LMG18311 were available. However, these models failed to represent the shift from mixed acid production to lactate production. This particular behavior of LAB illustrated the need of using alternative objective functions and contraints based on more experimental data. Besides, in the above models, they did not integrate reactions of flavor formation and this is important for the analysis of flavor formation. In our project, we focused on the production of ethanol and it is necessary to have a model of complete metabolic pathways.

*Lactococcus lactis subsp.* cremoris MG1363 is typical model strain in industrial dairy fermentations. In the previous papers,* L.lactis* MG1363 is widely used in the production of ethanol. This model is the only one model of L.lactis MG1363 which is available. The selected *L.lactis* MG1363 model in our project is a genome-scale, constraint-based model. This model accounted for 518 genes, 754 reactions, and 650 metabolites. The amino acid catabolism coupled with the flavor-forming reactions at a level of detail was added in this model. Continuous culture experiments were designed and subsequently implemented in this model to estimate the energetic parameters for growth and non-growth-associated maintenance. Moreover, the carbon and nitrogen metabolism were also considered in this model. This selected model is valuable and suitable for investigating the production of ethanol in *L.lactis* MG1363 (Flahaut et al., 2013). This model was used for predicting the production of flavor diacetyl. After in-silico simulation and comparison of different strategies, they found that the yield was dependent on the cytochrome oxidase activity and the diacetyl production yield could reached at 0.9mol/mol glucose. This indicates that this model can be effectively used to predict the yield of the target product under certain conditions(Liu et al., 2016a).

Besides, we also use MEMOTE to evaluate this existing model. We found the annotations of metabolites, reactions and genes are all 100% present. The final score reached 85%

## 5. Discussion  <a class="anchor" id=" Discussion "></a>

In the current work, the ethanol metabolism of *Lactococcus lactis* has been studied by optimizing the carbon source, knocking out genes, knocking out genes based on code prediction, and performing up-regulation and down-regulation of related metabolic enzymes.

When we change the carbon source and find the best ethanol production rate. We found that the *Lactococcus lactis* in the model cant be grow when glucose replaced with other carbon source which indicted that this lactic acid bacteria is highly dependent in glucose. Also, when we change the import rate of glucose with predicted  minimum media imported glucose rate. The results are really amazing, lower glucose import rate for glucose which showed a higher ethanol production rate. And higher glucose import rate really lower ethanol production rate. This may because the glucose repression which inhibit the ethanol synthesis related enzymes. Glucose repression is the total lactic acid bacteria response to glucose, which includes profound changes in lactic acid bacteria metabolism and activation and repression of a multitude of genes (Santangelo, G.M., 2006). This may also because high amount of glucose input can be influence the ethanol tolerance of lactic acid bacteria and even inactivated several enzymes, our result are similar conclusion for yeast fermentation ethanol with the problem of glucose repression (Rodrussamee, N., Sattayawat, P. and Yamada, M., 2018). 

For the anerobic and aerobic fermentation, the result showed that with or without oxygen will not influence the ethanol and biomass for *L.lactis.* This result possibly due to the limitations of the programming environment. In ethanol production usually need a anaerobic conditions. This is because in the presence of oxygen, with the help of NOX Oxidase, the NAD+ will be generated to get the rodox balance without the NAD+ generated by ethanol synthesis. Also, the lactic acid bacteria will want to grow better the acetic acid will generated and produce mor e ATP for growth.


Besides, we deleted several genes, including three lactate dehydrogenase genes, pta, and als. The metabolic reactions involved in these genes produce different yields of lactic acid, acetic acid and acetoin. This strategy should result in a increase in the flow rate of the ethanol because the by-product pathway is knocked out. However, the result was different from our expected result. Knocking out LDH, PTA and ALS did not increase the flow rate of ethanol. 

## 6. Conclusion  <a class="anchor" id=" Conclusion"></a>
The GSM model is chosen based on the literature. The MEMOTE report of this model has a good score. The optimization of the existing GSM model i.e. the improvement of the ethanol yield was accomplished in this project by; i) Selection of carbon source and optimization of flow rate, ii) Investigating ethanol production rate under thee aerobic and anaerobic condition iii) Consulting the literature, finding out the genes related to metabolic branches and through knocking out to increase the flow rate of the main product (ethanol) , iv) Investigating new possible gene knock-outs through OptKnock, v) Predicting and analyzing overexpression targets through FVA and FSEOF analysis.  It seems that the most promising strategy for our cell factory is the optimization of carbon source and it’s flow rate.

This could potentially include further optimizing the existing GSM model by experimental data, and introducing new alcohol dehydrogenase or overexpressing the existing alcohol dehydrogenase in the model.
